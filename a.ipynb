```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("company_sales_data.csv")
df.head()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['month_number'], df['total_profit'], marker='o', color='b', linewidth=2)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Company Profit Per Month')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['month_number'], df['total_units'], marker='o', color='r', linewidth=2, linestyle='dotted')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.title('Total Units Sold Per Month')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,7))
for col in df.columns[1:-2]:  # skipping month, total_units, total_profit
    plt.plot(df['month_number'], df[col], marker='o', label=col)
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.title('Monthly Sales Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['month_number'], df['toothpaste'], color='m')
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales')
plt.title('Toothpaste Sales Per Month')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(df['month_number']-0.2, df['facecream'], width=0.4, label='Face Cream')
plt.bar(df['month_number']+0.2, df['facewash'], width=0.4, label='Face Wash')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.title('Face Cream and Face Wash Sales Per Month')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(df['month_number'], df['bathingsoap'], color='skyblue')
plt.xlabel('Month Number')
plt.ylabel('Bathing Soap Sales')
plt.title('Bathing Soap Sales Per Month')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['total_profit'], bins=10, color='orange', edgecolor='black')
plt.xlabel('Profit Range')
plt.ylabel('Frequency')
plt.title('Profit Range Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(df['month_number'], df['total_profit'], color='g')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit Per Month')
plt.show()
```

```python
product_columns = df.columns[1:-2]
total_sales = df[product_columns].sum()
plt.figure(figsize=(10,6))
plt.pie(total_sales, labels=product_columns, autopct='%1.1f%%')
plt.title('Last Year Sales Distribution by Product')
plt.show()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10,8))

ax[0].plot(df['month_number'], df['bathingsoap'], color='blue', marker='o')
ax[0].set_title('Bathing Soap Sales Data')
ax[0].set_xlabel('Month Number')
ax[0].set_ylabel('Sales Units')
ax[0].grid(True)


ax[1].plot(df['month_number'], df['facewash'], color='red', marker='o')
ax[1].set_title('Face Wash Sales Data')
ax[1].set_xlabel('Month Number')
ax[1].set_ylabel('Sales Units')
ax[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,7))
bar_width = 0.1
months = df['month_number']
for i, col in enumerate(product_columns):
    plt.bar(months + i*bar_width, df[col], width=bar_width, label=col)

plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.title('Monthly Sales Data by Product')
plt.legend()
plt.show()
```
